```python
import pandas as pd

file_path="/jupiter/LolStats/LolStats_2025.csv"
df=pd.read_csv(file_path)

df.head(1)
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=df,x="difficulty")
plt.title("Champion difficulty")
plt.ylabel("count")
plt.xlabel("Difficulty")
plt.show
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x="rangetype",order=df['rangetype'].value_counts().index)
plt.title("Range type")
plt.ylabel("count")
plt.xlabel("Range type")
plt.show
```

```python
df.sort_values('herotype')
plt.figure(figsize=(8,5))
sns.countplot(data=df,x="herotype",order=df['herotype'].value_counts().index)
plt.title("Hero type")
plt.ylabel("count")
plt.xlabel("Hero type")
plt.show
```

```python
df.sort_values('resource')
plt.figure(figsize=(12,7))
sns.countplot(data=df,x="resource",order=df['resource'].value_counts().index)
plt.title("Resource")
plt.ylabel("Count")
plt.xlabel("Resource")
plt.show
```

```python
df.sort_values('herotype')
plt.figure(figsize=(12,7))
sns.countplot(data=df,x="herotype",order=df['herotype'].value_counts().index)
plt.title("Resource")
plt.ylabel("Count")
plt.xlabel("Herotype")
plt.show
```

```python
df.sort_values('role')
plt.figure(figsize=(20,20))
plt.xticks(rotation=90)
sns.countplot(data=df,x="role",order=df['role'].value_counts().index)
plt.title("Role")
plt.ylabel("Count")
plt.xlabel("Role")
plt.show
```
