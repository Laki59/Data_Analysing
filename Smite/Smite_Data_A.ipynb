```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
smite_data = pd.read_csv("D:\python pypi\smite_items.csv")
```

```python
#Chart of all items with their names and prices
plt.figure(figsize=(12, 55))
sns.barplot(x='Cost', y='Item', data=smite_data, palette='YlGnBu')
plt.title('Item Prices')
plt.xlabel('Price')
plt.ylabel('Item Name')
plt.show()
```

```python
#Distribution of items from item type
plt.figure(figsize=(10, 6))
sns.countplot(data=smite_data, y='Item_Type', order=smite_data['Item_Type'].value_counts().index)
plt.title('Distribution of Item Types')
plt.xlabel('Count')
plt.ylabel('Item Type')
plt.show()
```

```python
#Number of items per tier
plt.figure(figsize=(10, 6))
sns.countplot(data=smite_data,x="Item_Tier",order=smite_data["Item_Tier"].value_counts().index)
plt.xlabel("Item tier")
plt.ylabel("Number of items")
plt.title("Number of tier items")
plt.show()
```

```python
#Distribution of item cost from items
plt.figure(figsize=(10, 6))
sns.histplot(data=smite_data, x='Cost', bins=20)
plt.title('Distribution of Item Costs')
plt.xlabel('Cost of the item')
plt.ylabel('Number of items')
plt.show()
```

```python
#Distribution of items from total cost
plt.figure(figsize=(10, 6))
sns.histplot(data=smite_data,x="Total_Cost",bins=20)
plt.xlabel("Cost of item")
plt.ylabel("Number of items")
plt.show()
```

```python
#Average item cost per tier
plt.figure(figsize=(10, 6))
average_cost_by_tier = smite_data.groupby('Item_Tier')['Cost'].mean().sort_values(ascending=False)
sns.barplot(x=average_cost_by_tier.index, y=average_cost_by_tier.values, palette='Set2')
plt.title('Average Cost by Item Tier')
plt.xlabel('Item Tier')
plt.ylabel('Average Cost')
plt.show()
```
